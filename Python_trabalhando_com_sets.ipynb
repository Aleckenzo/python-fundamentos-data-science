# **Trabalhando com Sets no Python**

Sets são uma estrutura de dados não ordenada que armazena elementos únicos.

🔹 Principais características:
 - Não possuem ordem (não têm índice)
 - Não permitem elementos repetidos
 - São mutáveis (podem ser alterados)
 - Ideais para remover duplicatas e fazer operações matemáticas
 - Os elementos devem ser imutáveis (int, float, string, tupla)
 - Não aceita lista, dict, set (são mutáveis)
 - É iterável
 - Permite operações matemáticas

## **Criação dos sets**

### **Utilizando as chaves `{ }`:**




```python
carros = {'Jetta Variant', 'Passat', 'Crossfox', 'Dodge Jorney'}

print(carros)
```

```python
# Não permite elementos repetidos

elementos_repetidos = {1, 2, 3, 3, 4}
print(elementos_repetidos)
```

```python
# Permite elementos de tipos diferentes

elementos_diferentes = {2017, "Onix", 37123.04, True}
print(elementos_diferentes)
```

```python
# É iterável

for i in carros:
  print(i)
```

### **Utilizando listas para criar sets e a função `set()`**

```python
lista_escolas = ['Data Science', 'Programação', 'Front-end',
                 'Mobile', 'DevOps', 'UX & Design', 'Inovação e Gestão']
```

set() é uma função construtora built-in do Python que cria um objeto (instância) da classe set.
Converte qualquer iterável (listas, tuplas, strings) em um conjunto, eliminando elementos repetidos.

```python
escolas_set = set(lista_escolas)

print(escolas_set)
```

## **Operações com Sets**


set é um tipo de dado que representa um conjunto matemático, com elementos únicos e permite operações como união, interseção, diferença e diferença simétrica.

Para esse exercício, vamos precisar criar conjuntos com os **acessórios** de alguns modelos de carros, são eles: **Passat**, **Crossfox** e **Jetta Variant**.

```python
acessorios_passat = {'Rodas de Liga', 'Travas Elétricas', 'Piloto Automático',
                     'Central Multimídia'}
print(acessorios_passat)
```

```python
acessorios_crossfox = {'Piloto Automático', 'Teto Panorâmico', '4 X 4',
                       'Central Multimídia'}
print(acessorios_crossfox)
```

```python
acessorios_jetta = {'Controle de Estabilidade', 'Câmbio Automático', 'Travas Elétricas',
                    'Rodas de Liga'}
print(acessorios_jetta)
```

### **Disjunção** (Diferença simétrica)

<img src='https://i.imgur.com/o2wsgPo.png' width = 50%>

O set.isdisjoint() é um método de sets que verifica se dois conjuntos não têm elementos em comum.

```python
set.isdisjoint(acessorios_passat, acessorios_crossfox)
```

```python
A = {1, 2, 3}
B = {4, 5, 6}
C = {3, 4, 5}

print(A.isdisjoint(B))  # True, não têm elementos em comum
print(A.isdisjoint(C))  # False, têm um elemento em comum
```

### **Interseção**

<img src='https://i.imgur.com/6W7bxU1.png' width = 50%>

O set.intersection() é um método de sets que retorna os elementos que dois conjuntos têm em comum.

```python
A = {1, 2, 3}
B = {2, 3, 4}
C = {3, 4, 5}

# Usando o método
print(set.intersection(A, C))
print(B.intersection(C))

# Usando o operador &
print(A & B)  # {2, 3}
```

```python
acessorios_passat & acessorios_crossfox
```

```python
acessorios_passat & acessorios_jetta
```

```python
set.intersection(acessorios_passat, acessorios_crossfox)
```

```python
set.intersection(acessorios_passat, acessorios_crossfox, acessorios_jetta) != set() # Não há elementos em comum
```

### **União**

<img src='https://i.imgur.com/C3MrKmK.png' width = 50%>

O set.union() é o método que retorna um novo set com todos os elementos de dois ou mais conjuntos, sem repetição

```python
acessorios_passat | acessorios_crossfox
```

```python
set.union(acessorios_passat, acessorios_crossfox, acessorios_jetta, {'Teto Solar'})
```

Resumo dos operadores:
- |  união (union())
- &  interseção (intersection())
- -- diferença (difference())
- ^  diferença simétrica / disjunção (symmetric_difference())

### **Perfomance com vários elementos**

```python
elementos_set = set(range(10000))
elementos_list = list(range(10000))
elementos_tuple = tuple(range(10000))
```

```python
def verificar_elemento_in(elemento_iteravel):

  for i in range(10000):
    if i in elemento_iteravel:
      pass
```

```python
print("Tempo de execução set: ")
%time verificar_elemento_in(elementos_set)
print("="*30)

print("Tempo de execução lista: ")
%time verificar_elemento_in(elementos_list)
print("="*30)

print("Tempo de execução tuplas: ")
%time verificar_elemento_in(elementos_tuple)
print("="*30)
```
